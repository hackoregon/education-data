```python
import pandas as pd
```

```python
# Read in poverty data
poverty = pd.read_csv('../data/FreeReducedLunch.csv')
poverty
```

```python
# Select columns of interest
poverty = poverty[['SchoolID','Enrollment','Free_Lunches','Reduced_Lunches','Eligible_Students','PercentEligible']]
poverty
```

```python
# Eliminate incomplete cases
poverty = poverty.dropna()
poverty
```

```python
# Read in school data
schools = pd.read_csv('../data/schools.csv')
schools
```

```python
# Drop row 613/Chief Joseph
schools.drop_duplicates()
```

```python
# Read in SchoolType data
types = pd.read_csv('../data/RCmediaSchoolsAggregate.csv')
types
```

```python
# Select interesting columns
types = types[['SchoolID', 'SchoolType']]
types
```

```python
# Merge the collected data into one data frame
poverty_by_type = pd.merge(poverty[['SchoolID', 'Enrollment', 'Eligible_Students']], schools)
poverty_by_type = poverty_by_type.drop_duplicates()
poverty_by_type
```

```python
poverty_by_type = pd.merge(poverty_by_type, types)
poverty_by_type
```

```python
# Add together the Enrollment and EligibleStudents data for all rows
# that have the same District and SchoolType, thus giving poverty
# totals for each District across SchoolTypes (pandas is also great)
poverty_by_type = poverty_by_type.groupby(['District', 'SchoolType']).sum()
poverty_by_type
```

```python
# Compute a percentage in poverty for each District + SchoolType
poverty_by_type['poverty'] = poverty_by_type.Eligible_Students / poverty_by_type.Enrollment
poverty_by_type
```

```python
# Subset the columns to what we still care about
poverty_by_type = poverty_by_type['poverty']
poverty_by_type
```

```python
# Note that reducing a pandas data frame to one column
# makes it a Series (data frames are collections of Series)
type(poverty_by_type)
```

```python
# Instead of three rows for each District (one for each SchoolType),
# get one row for each District and three poverty columns
# (pandas don't need no reshape2 library overwrite)
poverty_by_type = poverty_by_type.unstack(level=-1)
poverty_by_type
```

```python
# Drop Districts for which we don't have at least one of each SchoolType
poverty_by_type = poverty_by_type.dropna()
poverty_by_type
```

```python
# Compute difference in poverty between High School and Elementary
# within Districts--is it negative as we suspect?
poverty_by_type['EH'] = poverty_by_type.H - poverty_by_type.E
poverty_by_type
```

```python
%matplotlib inline

poverty_by_type.EH.plot(kind='hist')
```
